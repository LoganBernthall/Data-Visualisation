```python
#Python Imports
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Importing CSV for traffic by region

csvdf = pd.read_csv('region_traffic_by_road_type.csv')

#Limit check - just seeing if df loaded correctly
result = csvdf.head(20)
print(result)
```

```python
#Data cleaning
#Check for null columns
print(csvdf.isnull().sum())
print('\n')#New line to make it easier to see
print(csvdf.isnull().any())
```

```python
list(csvdf.columns.values)
```

```python
#Data Visualisation - visualising road cataegory by name
#We do not need all the columns all the time, we will make multiple line graphs with different parts of data
cbmdf = csvdf.drop(['year','region_id','region_ons_code','road_category_id','link_length_km','region_name','link_length_miles','all_motor_vehicles'], axis=1)
print(cbmdf)
print('\n')#New line to make it easier to see

#x-axis
x = np.array(['TM', 'TA', 'PA','M', 'PM'])

#y-axis
cbmdf['road_category_name_count'] = cbmdf.groupby('road_category_name')['road_category_name'].transform('count')
CategCount = cbmdf.groupby('road_category_name')['road_category_name_count'].sum().reset_index()


#creating chart

plt.subplots(figsize=(60,60))
plt.xlabel('Road Category Name')
plt.xticks(rotation=90, fontsize=30)
plt.ylabel('Count', fontsize=30)
plt.bar(x, CategCount['road_category_name_count'])
plt.show()
```

```python
#Data Visualisation - visualising ammount of cars on M road traffic categoried roads in the southwest by year

#Creating a trimmed dataframe
swmdf = csvdf.drop(['region_id', 'region_ons_code', 'road_category_id','link_length_km','link_length_miles'], axis=1)
#print(swmdf) #Test print
trimdf = swmdf['region_name'] != 'South West' #Further trimming the df = this cuts all values which do not equal south west in the region name
swmdf = swmdf[~trimdf]
#print(swmdf)

x = np.array(['1993','1994','1995','1996','1997','1998','1999','2000','2001','2002','2003','2004','2005','2006','2007','2008','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019','2020','2021','2022','2023'])

y = swmdf.query('road_category_name == "M"')
y = y.drop(['year','region_name','road_category_name'], axis=1)

#print(y)

#Plotting
plt.subplots(figsize=(20,20))
plt.xlabel('Year')
plt.xticks(rotation=90, fontsize=30)
plt.yticks(fontsize=30)
plt.ylabel('Ammount Of Cars in millions', fontsize=30)
plt.grid(True)
plt.plot(x,y,color = 'red',linewidth=11)
plt.show()
```
