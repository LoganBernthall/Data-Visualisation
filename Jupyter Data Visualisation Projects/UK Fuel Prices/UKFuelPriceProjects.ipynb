```python
#Python Imports
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Importing CSV data for fuel prices
df = pd.read_csv('WeeklyFuelPricesUK.csv')

#Limiting current printed rows
result = df.head(10)

print(result) #Test print

```

```python
#-------------------------------------------------------------------------------------Data CLeaning-----------------------------------------------------------------------------
```

```python
#Data Cleaning - Some rows missing data which is unhelpful so we will remove
print(df.isnull().sum()) #This method shows what columns have empty values and how many
```

```python
#Data Cleaning - Some rows missing data which is unhelpful so we will remove --BINARY EQUIV OF PREVIOUS CELL
print(df.isnull().any())
```

```python
#Removing empty cells
clndf = df.dropna()

#Checking if there are any empty cells after this
print(clndf.isnull().sum())
```

```python
#Dropping columns which are not needed
#We don't need a lot of these columns as we are just visualising the trend of fuel prices
clndf.drop(['ULSP:  Diff on previous WEEK (p/litre)',' ULSP: Diff on previous  YEAR (p/litre)','Duty rate ULSP (p/litre)', 'VAT (% rate) ULSP', 'ULSD: Diff on previous WEEK (p/litre)','ULSD: Diff on previous  YEAR (p/litre)', 'Duty rate ULSD (p/litre)','VAT (% rate) ULSD'], axis=1)
```

```python
#-------------------------------------------------------------------------------------Data Visulisation-----------------------------------------------------------------------------
```

```python

clndf = pd.DataFrame(clndf, columns=["Date", "ULSP: Pump price (p/litre)", "ULSD: Pump price (p/litre)"])
clndf.plot(x='Date', y=['ULSD: Pump price (p/litre)', 'ULSP: Pump price (p/litre)'], figsize=(20, 20))

#plt.plot(clndf['Date'], clndf['ULSP: Pump price (p/litre)'],clndf['ULSD: Pump price (p/litre)'])

# Plotting the data

plt.xlabel('Date')
plt.ylabel('Pump price (p/litre)')
plt.title('Diesel Prices Over Time')
plt.grid(True)
plt.show()
```
